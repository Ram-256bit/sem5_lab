```python
 import nltk
```

```python
nltk.download('brown')
```

```python
from nltk.corpus import brown
print ('Total Categories :', len(brown.categories()))
print (brown.categories())
```

```python
sentences = brown.sents(categories='adventure')
print(sentences)
```

```python
tagged_words = brown.tagged_words(categories='adventure')
print(tagged_words)
```

```python
nouns = [(word, tag) for word, tag in tagged_words if any(noun_tag in tag for noun_tag in
['NP','NN'])]
```

```python
print(nouns[0:10])
```

```python
nouns_freq = nltk.FreqDist([word for word, tag in nouns])
```

```python
print(nouns_freq.most_common(10))
```

```python
adverbs = [(word, tag) for word, tag in tagged_words if tag.startswith('RB')]
print(adverbs[:10])
```

```python
adjectives = [(word, tag) for word, tag in tagged_words if tag.startswith('JJ')]
print(adjectives[:10])
```

```python
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
word = 'hike';
```

```python
word_synsets = wn.synsets(word)
print(word_synsets)
```

```python
for synset in word_synsets:

  print('Synset Name:', synset.name())
  print('POS Tag', synset.pos())
  print('Defenition', synset.definition())
  print('Examples', synset.examples())
```

```python
synset = wn.synsets(word)[0]
hypernyms = synset.hypernyms()
hyponyms = synset.hyponyms()
```

```python
print('Hypernyms:')
for hypernym in hypernyms:
  print(f'{hypernym.name()}: {hypernym.definition()}')
```

```python
print('\nHyponyms:')
for hyponym in hyponyms:
    print(f'{hyponym.name()}: {hyponym.definition()}')
```

```python
word1 = 'hike'
word2 = 'climb'
synsets1 = wn.synsets(word1)
synsets2 = wn.synsets(word2)
```

```python
print(f'Similarity scores between "{word1}" and "{word2}":')
for synset1 in synsets1:
    for synset2 in synsets2:
        similarity = synset1.wup_similarity(synset2)
        if similarity is not None:
            print(f'Synset1: {synset1.name()}, Synset2: {synset2.name()}, Similarity: {similarity}')
```
